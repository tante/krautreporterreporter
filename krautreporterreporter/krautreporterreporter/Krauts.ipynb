```python
import json
raw = json.load(open("/home/tante/krauter.json","r"))
data = [{"name":e['name'],'amount':float(e['amount'].replace(".","").replace(",","."))} for e in raw]
```

```python
data = sorted(data, key=lambda tup: tup['amount'], reverse=True)
```

```python
# Gesamtsumme
summe = 0
for spende in data:
    summe += spende['amount']
summe
```

```python
grossspender = [spender for spender in data if spender['amount']>500]
summe = 0
for s in grossspender:
    summe += s['amount']
# Geld aus Spenden > 500 €
summe
```

```python
grossspender = [spender for spender in data if spender['amount']>1000]
summe = 0
for s in grossspender:
    summe += s['amount']
# Geld aus Spenden > 1000
summe
```

```python
seen = []
seen_more_than_once = set()
for spender in data:
    if spender['name'] in seen:
        seen_more_than_once.add(spender['name'])
    if spender['name'] not in seen:
        seen.append(spender['name'])    

# Namen, die mehrfach spendeten
mehrfachspender = {}
for spende in seen_more_than_once:
    mehrfachspender[spende] = {'anzahl':0,'gesamt':0}

for spende in data:
    if spende['name'] in mehrfachspender.keys():
        mehrfachspender[spende['name']]['anzahl']+= 1;
        mehrfachspender[spende['name']]['gesamt']+= spende['amount']
        # what a lousy hack
        mehrfachspender[spende['name']]['name'] = spende['name']

# Mehrfachspender nach Häufigkeit
sorted(mehrfachspender.values(), key=lambda tup: tup['anzahl'], reverse=True)
```
